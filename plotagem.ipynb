```python
# Ajustando o espaamento em torno das subplotagens
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5) # axes coloca os valores
plt.subplots_adjust(wspace=0, hspace=0)
```
