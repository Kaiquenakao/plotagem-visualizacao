```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
data = pd.read_csv('examples/spx.csv', index_col=0, parse_dates=True)
spx = data['SPX']
spx.plot(ax=ax, style='k-')
crisis_data = [
    (datetime(2007, 10, 11), 'Peek of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='red', headwidth=4, width=2,
                               headlength=4),
                            horizontalalignment='left', verticalalignment='top')
    
# Faz um zoom no período de 2007 a 2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
#remover depois

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
spx = data['SPX']
spx.plot(ax=ax, style='k')

crisis_data = [
    (datetime(2007, 10, 11), 'Peek of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
               xytext=(date, spx.asof(date)+ 200),
               arrowprops=dict(facecolor='red', headwidth=4, width=1, 
                          headlength=10),
               horizontalalignment='left', verticalalignment='top')


ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
# Patches: objetos que representam muitas froams comuns
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.1, color='k', alpha=0.3) # ponta inicial da diagonal esquerdo inferior
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3) # começo da ponta do circulo e o outro é o raio
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                  color='g', alpha=0.5) # a ponta inferior esquerdo
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
macro = pd.read_csv('examples/macrodata.csv')
macro.head()
```

```python
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
data.head()
```

```python
trans_data = np.log(data).diff().dropna()
```

```python
trans_data
```

```python
trans_data[-5:]
```

```python
sns.regplot('m1', 'unemp', data=trans_data)
plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))
```

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})
```
